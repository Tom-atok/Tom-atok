# Seabornを用いた統計量の可視化チートシート（Google Colaboratory対応）

このチートシートでは、Google Colaboratory上で実践的にSeabornを使った基本的な統計量の可視化を学ぶことができます。以下に代表的な統計量の図示を行うためのコード例を、サンプルデータセットを用いてまとめました。



### 1. 基本的なインポートと設定
まず、必要なライブラリのインポートとテーマの設定を行います。


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Google Colab用のファイル操作ライブラリ
from google.colab import files

# テーマの設定
sns.set_theme(style="whitegrid")
```

### 2. サンプルデータの読み込み
Pandasを使って、Seabornに組み込まれているサンプルデータセットを読み込みます。


```python
# サンプルデータセットの読み込み
# 'tips' データセットはレストランでのチップに関する情報を含みます。
df = sns.load_dataset('tips')

# データの先頭5行を表示
print(df.head())
```

### 3. ヒストグラム
データの分布を視覚化するためのヒストグラムです。
- `bins`はビンの数を指定します。
- `kde=True`でカーネル密度推定を追加できます。

```python
# ヒストグラムの描画
sns.histplot(data=df, x="total_bill", bins=30, kde=True)
plt.show()
```

### 4. ボックスプロット
データの中央値や四分位範囲を視覚化するためのボックスプロットです。
- カテゴリごとのデータの分布を比較するのに便利です。


```python
# ボックスプロットの描画
sns.boxplot(data=df, x="day", y="total_bill")
plt.show()

```

### 5. バイオリンプロット
ボックスプロットに加えてデータの分布形状も示すバイオリンプロットです。
- `inner="quartile"`で四分位数を表示できます。



```python
# バイオリンプロットの描画
sns.violinplot(data=df, x="day", y="total_bill", inner="quartile")
plt.show()

```

### 6. ペアプロット
複数の変数間の関係を視覚化するためのペアプロットです。
- `hue`でカテゴリごとに色分けできます。


```python
# ペアプロットの描画
sns.pairplot(df, hue="sex")
plt.show()
```

### 7. 散布図と回帰直線
2つの変数間の関係を示す散布図と回帰直線の描画です。
- `hue`を指定すると、カテゴリごとに色分けされます。



```python
# 散布図と回帰直線の描画
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex")
plt.show()
```

### 8. ヒートマップ
相関行列などを視覚化するためのヒートマップです。
- `annot=True`で各セルに相関係数を表示します。



```python
# 数値データのみを選択して相関行列を計算
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

### 9. カウントプロット
カテゴリ変数のカウントを視覚化するカウントプロットです。
- 各カテゴリの出現頻度を示します。


```python
# カウントプロットの描画
sns.countplot(data=df, x="day")
plt.show()

```
